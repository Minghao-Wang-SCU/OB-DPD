```python
from rdkit import Chem
from rdkit.Chem import AllChem
from tqdm import  tqdm
import subprocess
import random
```

```python

def smi2mol(smi,name):
    

    # 将 SMILES 转换为 Mol 对象
    mol = Chem.MolFromSmiles(smi)

    # 添加氢原子并生成三维坐标
    mol_with_h = AllChem.AddHs(mol)
    #AllChem.EmbedMolecule(mol_with_h)
    AllChem.MMFFOptimizeMolecule(mol_with_h)

    # 保存为 .mol 文件
    Chem.MolToMolFile(mol_with_h, f'{name}.mol')

from rdkit import Chem
from rdkit.Chem import AllChem
def smiles2pdb(smiles,name):
    # 从 SMILES 生成分子对象
    mol = Chem.MolFromSmiles(smiles)
    # 添加氢原子
    mol_with_h = Chem.AddHs(mol)
    # 生成 3D 构象
    AllChem.EmbedMolecule(mol_with_h)
    # 创建 PDB 文件写入器
    writer = Chem.PDBWriter(f'{name}.pdb')
    # 将分子写入 PDB 文件
    writer.write(mol_with_h)
    # 关闭写入器
    writer.close()
smiesl_list = []
with open('GDB17.txt', 'r', encoding='utf-8') as file:
    for line in tqdm(file):
        smiesl_list.append(line.strip())
random_smiesl_list = random.sample(smiesl_list, 10000)
```
